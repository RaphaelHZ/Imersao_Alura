# DIA 02



```python
dados.columns = dados.columns.str.replace('droga','composto')
dados.head(5)
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
dados.head()
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(5,4))
plt.title('TOP 5 compostos')
plt.xlabel('Quantitade')
sns.countplot(y = 'composto', data=dados.query('composto in @cod_compostos'))
```

## DESAFIO 01: Ordenar o gr√°fico countplot

```python
cod_compostos = dados['composto'].value_counts().sort_values(ascending=False).index[:5]
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=dados.query('composto in @cod_compostos'), x='composto', order=cod_compostos)
ax.set_title("Top 5 compostos")
ax.set_xlabel('Compostos')
ax.set_ylabel('Frequ√™ncia')
plt.show()
```
